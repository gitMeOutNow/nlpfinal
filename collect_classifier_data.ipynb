```python
classfication_data = {"class": [], "sentence": []}

topics = ['chemistry', 'economics', 'polisci', 'psychology', 'sociology']

for topic in topics:
  print(f'---- TOPIC => {topic} ----')
  directory = f'/content/gdrive/MyDrive/data/metadata/{topic}'

  for file in os.listdir(directory):
    print(f'adding file {file}')
    file_path = directory + f'/{file}'
    for i, line in enumerate(open(file_path, 'r')):
      doc = json.loads(line)
      text = doc['title']

      tokenizer = RegexpTokenizer(r'\w+')
      sentences = nltk.tokenize.sent_tokenize(text)
      sentences = [" ".join(tokenizer.tokenize(sentence.lower())) for sentence in sentences]
      sentences = [s + '\n' for s in sentences]

      classfication_data['class'].append(topic)

      sentence = sentences[-1]
      classfication_data['sentence'].append(sentence)

      print(sentence)

      if i > 10000:
        break

      break

    break

  break
      
# df = pd.DataFrame(classfication_data)
# df.to_csv(f'/content/gdrive/MyDrive/data/classification_data/classification_train.tsv', sep = " ")
```
