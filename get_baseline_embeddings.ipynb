```python
import gensim
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
```

```python
%cd /content/gdrive/MyDrive/data/embeddings
```

```python
genres = ['polisci', 'economics', 'psychology', 'sociology', 'chemistry']
dim = 100
for genre in genres:
  src_loc = genre  + '_wordvec.model'
  model = KeyedVectors.load(src_loc, mmap='r')
  save_dest = 'word2vec.' + genre + '.' + str(dim) + 'd.txt'
  model.wv.save_word2vec_format(save_dest, binary=False)
```
