```python
import json
import pprint
import os
import pandas as pd
import nltk
from nltk.tokenize import RegexpTokenizer
# from rake_nltk import Rake
from gensim.summarization import keywords

nltk.download('stopwords')
# nltk.download('wordnet')
nltk.download('punkt')
```

```python
%cd /content/gdrive/MyDrive/data
%pwd
```

```python
TOPIC = "chemistry"
INPUT_DIRECTORY = f'/content/gdrive/MyDrive/data/text_data/{TOPIC}'
OUTPUT_DIRECTORY = f'/content/gdrive/MyDrive/data/eval_data/masked_tokens/{TOPIC}'
```

```python
masked_data = {"masked_token": [], "masked_line": [], "masked_idx": []}
for file in os.listdir(INPUT_DIRECTORY):
  print(file)
  file_path = f'{INPUT_DIRECTORY}/{file}'
  for i, line in enumerate(open(file_path, 'r')):
    # create a mask of the sentence using the keyword
    extracted_keywords = keywords(line, split=True)
    
    if len(extracted_keywords) == 0:
      continue
    keyword = extracted_keywords[0]

    if " " in keyword:
      continue

    tokenizer = RegexpTokenizer(r'\w+')
    sentence = nltk.tokenize.word_tokenize(line)

    # cap max sentence length
    if len(sentence) > 50:
      continue

    try:
      keyword_index = sentence.index(keyword)
    except:
      continue

    masked_line = line.replace(keyword, "[MSK]", 1)
    masked_data["masked_token"].append(keyword)
    masked_data["masked_line"].append(masked_line)
    masked_data["masked_idx"].append(keyword_index)

    if i > 10000:
      break
  
df = pd.DataFrame(masked_data)
df.to_csv(f'{OUTPUT_DIRECTORY}/{TOPIC}.tsv', sep = " ")
```
