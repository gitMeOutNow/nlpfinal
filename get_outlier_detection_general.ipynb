```python
import json
import pprint
import os
import nltk
import pandas as pd
import random
from nltk.tokenize import RegexpTokenizer
nltk.download('punkt')
```

```python
topics = ['chemistry', 'economics', 'polisci', 'psychology', 'sociology']

classification_data = {"s0": [], "s1": [], "s2": [], "s3": [], "outlier": []}

MAX_COUNT = 20000

for topic in topics:
  file_path = f'/content/gdrive/MyDrive/data/eval_data/outlier_detection/{topic}/{topic}.tsv'
  df = pd.read_csv(file_path, sep=' ', header=0)
  df = df.sample(frac=1)
  for i in range(MAX_COUNT):
    classification_data["s0"].append(df.iloc[i]['s0'])
    classification_data['s1'].append(df.iloc[i]['s1'])
    classification_data['s2'].append(df.iloc[i]['s2'])
    classification_data['s3'].append(df.iloc[i]['s3'])
    classification_data['outlier'].append(df.iloc[i]['outlier'])

print(classification_data['s0'][0:10])
print(classification_data['s1'][0:10])
print(classification_data['s2'][0:10])
print(classification_data['s3'][0:10])
print(classification_data['outlier'][0:10])

df_output = pd.DataFrame(classification_data)
df_output.to_csv(f'/content/gdrive/MyDrive/data/eval_data/outlier_detection/general_data.tsv', sep = " ")
```
