```python

def each_chunk(stream, separator):
  buffer = ''
  while True:  # until EOF
    chunk = stream.read(4096) 
    if not chunk:  # EOF?
      yield buffer
      break
    buffer += chunk
    while True:  # until no separator is found
      try:
        part, buffer = buffer.split(separator, 1)
      except ValueError:
        break
      else:
        yield part

file_head = '/content/gdrive/MyDrive/data/text_data/'
class Corpus:
  def __init__(self, intro, max_num):
      self.file_tail = intro
      self.file_num = max_num
  def __iter__(self):
      
      for x in range(self.file_num):
        file_path = file_head + self.file_tail + str(x) + ".txt"
        corpus_path = datapath(file_path)
        with open(corpus_path) as opened_file:
          for line in each_chunk(opened_file, '\n'):
            yield utils.simple_preprocess(line)
      
```
