```python
import json
import pprint
import os
import nltk
import pandas as pd
import random
from nltk.tokenize import RegexpTokenizer
nltk.download('punkt')
```

```python
file_path = '/content/gdrive/MyDrive/data/pdf_parses/economics/pdf_parses_26.jsonl'

for line in open(file_path, 'r'):
  
  doc = json.loads(line)
  body_text = doc['body_text']

  if len(body_text) != 0:
    body_text = doc['body_text']
    text = ""
    for section in body_text:
      text += section["text"] 
    pprint.pprint(text)
    # print(text)
    break
```

```python
TOPIC = 'economics'
METADATA_DIRECTORY = f'/content/gdrive/MyDrive/data/metadata/{TOPIC}'

query_title_data = {"title": [], "abstract_sentence": [], "category": []}
title_data = {"title": [], "category": []}
titles = []

total_count = 0
for file in os.listdir(METADATA_DIRECTORY):
  print(f'adding file {file}')
  metadata_path = METADATA_DIRECTORY + f'/{file}'  
  for i, line in enumerate(open(metadata_path, 'r')):
    doc = json.loads(line)
    title = doc['title'].lower()
    abstract = doc['abstract']
    if abstract == None:
      continue
    
    tokenizer = RegexpTokenizer(r'\w+')
    sentences = nltk.tokenize.sent_tokenize(abstract)
    sentences = [" ".join(tokenizer.tokenize(sentence.lower())) for sentence in sentences]
    sentences = [s + '\n' for s in sentences]

    split_len = len(sentences[0].split(" "))
    if split_len > 5 and split_len < 35:
      query_title_data["title"].append(title)
      query_title_data['abstract_sentence'].append(sentences[0])
      query_title_data['category'].append(TOPIC)

      title_data["title"].append(title)
      title_data['category'].append(TOPIC)

      titles.append(title)

      total_count += 1

      if i > 1000:
        break

  print(total_count)
  if total_count > 5000:
    break

df_query_title_data = pd.DataFrame(query_title_data)
df_title_data = pd.DataFrame(title_data)

print(df_query_title_data.head())
print(df_title_data.head())

df_query_title_data.to_csv(f'/content/gdrive/MyDrive/data/eval_data/top_k_queries/{TOPIC}/{TOPIC}_query_title.tsv', sep = " ")
df_title_data.to_csv(f'/content/gdrive/MyDrive/data/eval_data/top_k_queries/{TOPIC}/{TOPIC}_title.tsv', sep = " ")
```
