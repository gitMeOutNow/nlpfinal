```python
from gensim.models.keyedvectors import KeyedVectors

model = KeyedVectors.load_word2vec_format('/content/gdrive/MyDrive/embeddings/word2vec/baseline/word2vec_embeddings.bin', binary=True)
model.save_word2vec_format('/content/gdrive/MyDrive/embeddings/word2vec/baseline/word2vec_embeddings.txt', binary=False)
```

```python
"""
chemistry -- DONE
economics -- DONE
polisci -- DONE
sociology - DONE
psychology -- DONE
"""
TOPIC = 'psychology'
category_vocab_set = set()
CATEGORY_FILE = f'/content/gdrive/MyDrive/embeddings/word2vec/word2vec.{TOPIC}.100d.txt'
for i, line in enumerate(open(CATEGORY_FILE, 'r')):
  if i == 0:
    continue
  else:
    word = line.split(" ")[0]
    category_vocab_set.add(word)

  if i % 50000 == 0:
    print(f'done with {i} words')

print(len(category_vocab_set))
```

```python
output_file = open(f'/content/gdrive/MyDrive/embeddings/word2vec/baseline/word2vec_embeddings_{TOPIC}.txt', 'w')
BASELINE_FILE = '/content/gdrive/MyDrive/embeddings/word2vec/baseline/word2vec_embeddings.txt'
j = 0
for i, line in enumerate(open(BASELINE_FILE, 'r')):
  word = line.split(" ")[0].lower()
  if word in category_vocab_set:
    output_file.write(line.lower())
    category_vocab_set.remove(word)
    j += 1

  if i % 10000 == 0:
    print(f'done with {i} words (j = {j})')
```
